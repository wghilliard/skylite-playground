# claudia criterion

this notebook illustrates a criterion for dropping overly dark/gray colors from the palette invented by my girlfriend.

```python
from PIL import Image
import numpy as np
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
hex_color = Image.open('./data/training/hex_color.jpg')
h_c_shape = tuple(reversed((hex_color.size)))
```

```python
X = np.array(hex_color).reshape((-1, 3))
```

```python
def f0(r, g, b):
    return .2*(255 + 255 + 255 - r - g - b)/3

def f1(r, g, b):
    return max([r, g, b]) - min([r, g, b])

def f2(r, g, b):
    return 1 if f0(r, g, b) >= f1(r, g, b) else 0
```

```python
y = np.array([f2(x[0], x[1], x[2]) for x in X]).reshape(h_c_shape)
```
